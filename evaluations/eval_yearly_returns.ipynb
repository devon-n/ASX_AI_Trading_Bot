```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Turn this into a function with params

# Read Data
with open('./yearly_returns.txt', 'r') as f:
    returns = f.readlines()
returns = eval(returns[0])
x = [i for i in range (len(returns))]

# Above and below
below = [i for i in returns if i < 10]
above = [i for i in returns if i >= 10]

print("Below 10%: ", round((len(below) / len(returns)) * 100), "%")
print("Above 10%: ", round((len(above) / len(returns)) * 100), "%")
print()

# Describe
df = pd.Series(returns)
print('Median: ',df.median())
print()
print(df.describe())
print()

# Line Chart
plt.figure(figsize=(26,7))
plt.plot(x, returns)
plt.show()

# Histogram
fig, ax = plt.subplots(figsize=(26,7), tight_layout=True)
ax.hist(returns, bins=[-20,-10,-8,-6,-4,-2,0,2,4,6,8,10,20])
ax.set_xticks(range(-20, 20))
plt.show()
```
